## $\mathbb{C}$-linearity

**Definition.**  
Let $V, W$ be two vector spaces over the real number field $\mathbb{R}$.  
A mapping
$T: V \to W$
is said to be __$\mathbb{R}$-linear__ if

- $T(cx) = cT(x) \quad\forall\; c \in \mathbb{R}, x \in V$
- $T(x + y) = T(x) + T(y) \quad\forall\; x, y \in V$

**Definition.** (Complex scalar multiplication of an $\mathbb{R}^{2}$ vector)  
For all
$$
w \in \mathbb{C},
\begin{pmatrix}
a \\
b
\end{pmatrix}
\in \mathbb{R}^{2},
$$

we define

$$
w
\begin{pmatrix}
a \\
b
\end{pmatrix}
:=
\begin{pmatrix}
\Re(w) \\
\Im(w)
\end{pmatrix}
\begin{pmatrix}
a \\
b
\end{pmatrix}
=
\begin{pmatrix}
\Re(w)a - \Im(w)b \\
\Re(w)b + \Im(w)a
\end{pmatrix}.
$$

**Definition.**  
A $\mathbb{R}$-linear mapping
$T: \mathbb{R}^{2} \to \mathbb{R}^{2}$
is said to be __$\mathbb{C}$-linear__ if

$$
T(wx) = wT(x) \quad\forall\; w \in \mathbb{C}, x \in \mathbb{R}^{2}.
$$

**Example.**  
The $\mathbb{R}$-linear mapping induced by the real matrix
$\begin{pmatrix} a & b \\ c & d\end{pmatrix}$

$$
\begin{pmatrix}
x \\
y
\end{pmatrix} \mapsto
\begin{pmatrix}
a & b \\
c & d
\end{pmatrix}
\begin{pmatrix}
x \\
y
\end{pmatrix}
$$

might give the **illusion** that it could be **$\mathbb{C}$-linear as well**.

But, being careful enough, one finds out quickly that **it's not**, e.g.

$$
\begin{pmatrix}
a & b \\
c & d
\end{pmatrix} \left(i
\begin{pmatrix}
x \\
y
\end{pmatrix}\right)
=
\begin{pmatrix}
a & b \\
c & d
\end{pmatrix}
\begin{pmatrix}
-y \\
x
\end{pmatrix}
=
\begin{pmatrix}
-ay + bx \\
-cy + dx
\end{pmatrix}
$$

while

$$
i \left( \begin{pmatrix}
a & b \\
c & d
\end{pmatrix}
\begin{pmatrix}
x \\
y
\end{pmatrix} \right)
=
i
\begin{pmatrix}
ax + by \\
cx + dy
\end{pmatrix}
=
\begin{pmatrix}
-cx - dy \\
ax + by
\end{pmatrix}.
$$

**Rmk.**  
From the above, we could also see that a necessary condition
for the induced $\mathbb{R}$-linear mapping to be
$\mathbb{C}$-linear mapping is

$$
\begin{cases}
c = -b \\
d = a
\end{cases}\;.
$$

Writing a complex-valued function of a complex variable $f: \mathbb{C} \to \mathbb{C}$ as

$$f = P(x,y) + iQ(x,y)\,,$$

we see that the real matrix above

$$
\begin{pmatrix}
a & b \\
c & d
\end{pmatrix}
=
\begin{pmatrix}
P_{x} & P_{y} \\
Q_{x} & Q_{y}
\end{pmatrix}
$$

and the condition

$$
\begin{cases}
c = -b \\
d = a
\end{cases}
$$

is none other than the Cauchy-Riemann equations

$$
\begin{cases}
Q_{x} = -P_{y} \\
Q_{y} = P_{x}
\end{cases}\,.
$$

**Rmk.**  
> _Complex-differentiable functions are precisely
> real-differentiable functions whose differential
> happens to be $\mathbb{C}$-linear._
